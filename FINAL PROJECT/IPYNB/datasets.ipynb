```python
#MERGE IMPORT AND EXPORT TOGETHER 

year_columns = ['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

#add a prefix to the year columns for both DataFrames
Exports.columns = ['Country'] + [f'Exports_{year}' for year in year_columns]
Imports.columns = ['Country'] + [f'Imports_{year}' for year in year_columns]

#remove duplicates from 'Exports' and 'Imports' based on the 'Country' column
Exports = Exports.drop_duplicates(subset='Country')
Imports = Imports.drop_duplicates(subset='Country')

#merge
merge_import_export = pd.merge(Exports, Imports, on='Country')


merge_import_export.head(50)
```
