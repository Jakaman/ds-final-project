```python
import pandas as pd 

df = pd.read_excel("../data/Data_FINAL_PROJECT.xlsx")
```

```python
#remove mapped_country because it is useless now 

df = df.drop(columns=['Mapped_Country'])

df = df.dropna()

df.columns
```

```python
#convert all columns to numeric EXCEPT COUNTRY
df = df.drop(index=0)

df = df.dropna()

for col in df:
    if col !='Country':
        
        df[col] = pd.to_numeric(df[col], errors='coerce')
        

df.head()
```

```python
rows_with_nulls = df[df.isnull().any(axis=1)]

# Display the first few rows that contain null values
rows_with_nulls.head()
```

```python
df_cleaned = df.dropna()

rows_with_nulls = df_cleaned[df_cleaned.isnull().any(axis=1)]

rows_with_nulls

df_cleaned.head(50)
```

```python
#Convert PIB into Millions USD 

for column in df_cleaned:
    if column.startswith('GDP_'):
        df_cleaned[column] = df_cleaned[column].astype(float) / 1e6
```

```python
#Create Trade Openness / Openness Ratio variable

for year in range(2014, 2023):
    export_column = f'Exports_{year}'
    import_column = f'Imports_{year}'
    gdp_column = f'GDP_{year}'
    openness_column = f'Openness_{year}'

    df_cleaned[export_column] = df_cleaned[export_column].astype(float)
    df_cleaned[import_column] = df_cleaned[import_column].astype(float)

    #Trade Openness formula
    df_cleaned[openness_column] = (df_cleaned[export_column] + df_cleaned[import_column]) / 2 / df_cleaned[gdp_column] * 100
```

```python
#drop inflation 

columns_to_drop = [col for col in df_cleaned.columns if 'Inflation' in col]
df_cleaned = df_cleaned.drop(columns=columns_to_drop)
```

```python
import pandas as pd

# Liste de DataFrames reprÃ©sentant chaque groupe de pays
Countries_group = [
    Europe_countries,
    middle_east_central_asia_countries,
    sub_saharan_africa_countries,
    western_hemisphere_countries,
    Asia_Countries
]

# Noms des groupes pour l'index du DataFrame final
group_names = [
    "Europe_countries",
    "middle_east_central_asia_countries",
    "sub_saharan_africa_countries",
    "western_hemisphere_countries",
    "Asia_Countries"
]

# CrÃ©ation d'un DataFrame vide pour stocker les moyennes de chaque groupe de pays
averages_by_group = pd.DataFrame()

# Boucle Ã  travers les DataFrames de chaque groupe de pays
for i, group_df in enumerate(Countries_group):
    # Calcul des moyennes par annÃ©e de 2014 Ã  2022
    averages = {}
    for year in range(2014, 2023):
        year_columns = [f'{col}_{year}' for col in ['GDP', 'Exports', 'Imports', 'Openness']]
        averages[f'Avg_GDP_{year}'] = group_df[year_columns[0]].mean()
        averages[f'Avg_Exports_{year}'] = group_df[year_columns[1]].mean()
        averages[f'Avg_Imports_{year}'] = group_df[year_columns[2]].mean()
        averages[f'Avg_Openness_{year}'] = group_df[year_columns[3]].mean()
    
    # CrÃ©er un DataFrame avec les moyennes pour ce groupe de pays
    df_averages_group = pd.DataFrame(averages, index=[group_names[i]])
    
    # Ajouter ce DataFrame au tableau gÃ©nÃ©ral
    averages_by_group = pd.concat([averages_by_group, df_averages_group])

# Afficher le tableau des moyennes par groupe de pays
print(averages_by_group)
```

```python
import plotly.graph_objects as go
import ipywidgets as widgets

# Exemple de donnÃ©es (remplacez ceci par vos vraies donnÃ©es)
regional_data = {
    'Asia': Asia_Countries,  # Remplacez par vos DataFrames
    'Sub-Saharan Africa': sub_saharan_africa_countries,
    'Middle East and Central Asia': middle_east_central_asia_countries,
    'Western Hemisphere': western_hemisphere_countries,
    'Europe': Europe_countries
}

# DÃ©finition des annÃ©es et des colonnes correspondantes pour les variables
years = ['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
export_columns = [f'Exports_{year}' for year in years]
import_columns = [f'Imports_{year}' for year in years]
gdp_columns = [f'GDP_{year}' for year in years]
openness_columns = [f'Openness_{year}' for year in years]

# Widgets pour la sÃ©lection du continent et de la variable
continent_selector = widgets.Dropdown(
    options=['Asia', 'Sub-Saharan Africa', 'Middle East and Central Asia', 'Western Hemisphere', 'Europe'],
    value='Asia',
    description='Continent:',
)

variable_selector = widgets.Dropdown(
    options=['Exports', 'Imports', 'GDP', 'Openness'],
    value='Exports',
    description='Variable:',
)

# Fonction pour mettre Ã  jour le graphique
def update_graph(continent, variable):
    region_df = regional_data[continent]
    if variable == 'Exports':
        y_data = region_df[export_columns].mean() / 1e6
    elif variable == 'Imports':
        y_data = region_df[import_columns].mean() / 1e6
    elif variable == 'GDP':
        y_data = region_df[gdp_columns].mean() / 1e6
    else:  # Openness
        y_data = region_df[openness_columns].mean()

    fig = go.Figure(data=go.Scatter(x=years, y=y_data, mode='lines+markers'))
    fig.update_layout(title=f'{variable} for {continent}', xaxis_title='Year', yaxis_title=variable)
    fig.show()

# Affichage interactif
widgets.interactive(update_graph, continent=continent_selector, variable=variable_selector)
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

def check_multicollinearity(X):
    # Cette fonction calcule le VIF pour chaque variable explicative
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

def perform_regression(df_region, year):
    # VÃ©rifiez si df_region est vide
    if df_region.empty:
        print(f"Aucune donnÃ©e disponible pour la rÃ©gion spÃ©cifiÃ©e en {year}")
        return

    # PrÃ©paration des donnÃ©es pour la rÃ©gression
    X = df_region[[f'Exports_{year}', f'Inflation_{year}']]
    X = sm.add_constant(X)
    y = df_region[f'GDP_{year}']

    # VÃ©rifiez si X ou y contient des valeurs manquantes
    if X.isnull().values.any() or y.isnull().values.any():
        print(f"DonnÃ©es insuffisantes pour la rÃ©gression en {year}")
        return

    # VÃ©rifier la multicolinÃ©aritÃ©
    vif_data = check_multicollinearity(X.drop(columns=['const'])) # exclure la constante du calcul du VIF
    print("VIF pour chaque variable:")
    print(vif_data)

    # Gestion de la multicolinÃ©aritÃ©
    while vif_data['VIF'].max() > 10:  # Seuil de VIF
        # Retirez la variable avec le VIF le plus Ã©levÃ©
        highest_vif = vif_data.sort_values('VIF', ascending=False).iloc[0]
        X = X.drop(columns=[highest_vif['feature']])
        vif_data = check_multicollinearity(X.drop(columns=['const']))
        print("VIF mis Ã  jour pour chaque variable:")
        print(vif_data)
    # Construction et affichage du modÃ¨le de rÃ©gression
    model = sm.OLS(y, X).fit()
    print(f"RÃ©sultats pour l'annÃ©e {year} dans la rÃ©gion spÃ©cifiÃ©e:")
    print(model.summary())
    print("\n")

# Exemple d'utilisation de la fonction
# Notez que vous devez dÃ©finir 'df_cleaned' avec vos donnÃ©es avant d'utiliser ce code
Europe_countries = df_cleaned.iloc[20:34]  # DonnÃ©es pour les pays europÃ©ens
Asia_Countries = df_cleaned.iloc[0:20]
middle_east_central_asia_countries = df_cleaned.iloc[34:57]
sub_saharan_africa_countries =df_cleaned.iloc[57:91] 
western_hemisphere_countries =df_cleaned.iloc[91:114] 
for year in range(2014, 2023):
    print(f"--- Analyse pour l'Asie en l'annÃ©e {year} ---")
    perform_regression(Asia_Countries, year)

    print(f"--- Analyse pour l'Europe en l'annÃ©e {year} ---")
    perform_regression(Europe_countries, year)
    
    print(f"--- Analyse pour le middle east en l'annÃ©e {year} ---")
    perform_regression(middle_east_central_asia_countries, year)

    print(f"--- Analyse pour l'Afrique en l'annÃ©e {year} ---")
    perform_regression(sub_saharan_africa_countries, year)

    print(f"--- Analyse pour le western en l'annÃ©e {year} ---")
    perform_regression(western_hemisphere_countries, year)


```
